```python
## Data Collection
import nltk
nltk.download("gutenberg")
from nltk.corpus import gutenberg
import pandas as pd
 
## Load the database
data=gutenberg.raw("shakespeare-hamlet.txt")

## Save to a file
with open("hamlet.txt","w") as file:
    file.write(data)
```

```python
### Data Preprocessing
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
```

```python
## Load the dataset
with open("hamlet.txt","r") as file:
    text=file.read().lower()
```

```python
###Tokenizing the text--creating index for words
tokenizer=Tokenizer()
tokenizer.fit_on_texts([text])
total_words=len(tokenizer.word_index)+1
total_words
```

```python
tokenizer.word_index
```

```python
## Creating input sequences
input_sequences=[]
for line in text.split('\n'):
    token_list=tokenizer.texts_to_sequences([line])[0]
    for i in range(2,len(token_list)):
        n_gram_sequence=token_list[:i+1]
        input_sequences.append(n_gram_sequence)

        
```

```python
input_sequences
```

```python
## Pad sequences
max_sequences_len=max(len(x) for x in input_sequences)
max_sequences_len
```

```python
input_sequences=np.array(pad_sequences(input_sequences,maxlen=max_sequences_len,padding="pre"))
input_sequences
```

```python
##Create predictors and labels
import tensorflow as tf
x=input_sequences[:,:-1]
y=input_sequences[:,-1]
x
y
```

```python
tf.keras.utils.to_categorical(y,num_classes=total_words)
```

```python
## Spliting data into training and testing
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
### Train LSTM RNN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout

## Define the model
model=Sequential()
model.add(Embedding(total_words,output_dim=100,input_length=max_sequences_len-1))
model.add(LSTM(150,return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(100))
model.add(Dense(total_words,activation="softmax"))

##Compiling the model
model.compile(loss="sparse_categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
model.build(input_shape=(None, max_sequences_len)) 
model.summary()
```

```python
## Train the code
history=model.fit(x_train,y_train,epochs=50,validation_data=(x_test,y_test),verbose=1)
```

```python
def predict_next_word(model,tokenizer,text,max_sequences_len):
    token_list= tokenizer.texts_to_sequences([text])[0]
    if len(token_list) >= max_sequences_len:
        token_list= token_list[-(max_sequences_len-1):]
    token_list= pad_sequences([token_list],maxlen=max_sequences_len-1,padding='pre')
    predicted=model.predict(token_list,verbose=0)
    predicted_word_index= np.argmax(predicted,axis=1)
    for word,index in tokenizer.word_index.items():
        if index== predicted_word_index:
            return word
    return None
```

```python
input_text="To be or not to be"
print(f"Input text; {input_text}")
max_sequences_len=model.input_shape[1]+1
next_word=predict_next_word(model,tokenizer,input_text,max_sequences_len)
print(f"Next predicted word:{next_word}")
```
